### create new column(date)

```python
fir_df["date"]=fir_df["FIR_Date"].str.split("/").str[0]
```

```python
fir_df["date"].unique()
```

```python
#converting into integer
fir_df["date"] = fir_df["date"].astype(int)
```

## District column

```python
fir_df["District_Name"].replace(['Belagavi City','Belagavi Dist'],'Belagavi',inplace=True)
fir_df["District_Name"].replace(['Bengaluru City','Bengaluru Dist','ISD Bengaluru','CID'],'Bengaluru',inplace=True)
fir_df["District_Name"].replace(['Hubballi Dharwad City'],'Dharwad',inplace=True)
fir_df["District_Name"].replace(['Kalaburagi City'],'Kalaburagi',inplace=True)
fir_df["District_Name"].replace(['K.G.F'],'Kolar',inplace=True)
fir_df["District_Name"].replace(['Mangaluru City'],'Mangaluru',inplace=True)
```

```python
fir_df["District_Name"].unique()
```

```python
fir_df["FIR Type"].unique() #crime type
```

Filling 'Not Applicable' in FIR TYPE

```python
fir_df["FIR Type"].fillna("Not Applicable",inplace=True)
fir_df.rename(columns={'FIR Type':'CrimeType'},inplace=True)
fir_df["CrimeType"].unique()
```

# Data Preparation

```python
label_encoder = preprocessing.LabelEncoder() 
fir_df['CrimeGroup_Name']= label_encoder.fit_transform(fir_df['CrimeGroup_Name']) 
```

```python
x = fir_df.iloc[:,[0,1,4,5,10,41]] #independent
y = fir_df.iloc[:,13]
```

```python
#encoding categorical data
x_d = pd.get_dummies(x,columns=['District_Name','UnitName','CrimeType'],dtype='int')
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_d,y,test_size=0.25,random_state=0)
```
